```python
import os
from dotenv import load_dotenv
from google.genai import types

load_dotenv("notebooks/.env")
```

```python
from tiny_agent.tools.web.tools import google_search
```

```python
_PROVIDER_CONFIG = {
    "vertexai": bool(os.environ.get("GOOGLE_GENAI_USE_VERTEXAI", True)),
    "vertexai_location": os.environ.get("GOOGLE_CLOUD_LOCATION", "europe-west4"),
    "vertexai_project": os.environ.get(
        "GOOGLE_CLOUD_PROJECT", "hg-hjghjg-ai-ft-exp-pr-hjjkhljhlhjkl"
    ),
    "google_ai_studio_api_key": os.environ.get("GOOGLE_AI_STUDIO_API_KEY", ""),
}

_MAIN_AGENT_MODEL = "gemini-3-flash-preview"
_MAIN_AGENT_MODEL_CONFIG = {
    "temperature": 1.0,
    "seed": 42,
    "top_p": 1.0,
    "top_k": 60,
    "thinking_config": types.ThinkingConfig(
        thinking_level=types.ThinkingLevel.HIGH,
        #thinking_budget=-1,
        include_thoughts=False,
    ),
}

_SUBAGENT_MODEL = "gemini-2.5-flash-lite"
_SUBAGENT_MODEL_CONFIG = {
    "temperature": 1.0,
    "seed": 42,
    "top_p": 1.0,
    "top_k": 60,
    "thinking_config": types.ThinkingConfig(
        #thinking_level=types.ThinkingLevel.HIGH,
        thinking_budget=-1,
        include_thoughts=False,
    ),
}

_PROVIDER_CONFIG, _MAIN_AGENT_MODEL, _MAIN_AGENT_MODEL_CONFIG,_SUBAGENT_MODEL,_SUBAGENT_MODEL_CONFIG
```

```python
from tiny_agent.agent.tiny_agent import TinyAgent, AgentManager 
```

```python
from tiny_agent.subagent.decorator import subagent

@subagent(is_async=True)
class WebResearchAgent(TinyAgent):
    """Perform research for a given topic on the web"""
    ... 
```

```python
num_members = 5
web_research_agents = [
    WebResearchAgent(
        name=f"web_research_{i}",
        model=_SUBAGENT_MODEL,
        output_root="notebooks/output",
        tools=[google_search],
        **{**_SUBAGENT_MODEL_CONFIG, **_PROVIDER_CONFIG},
    )
    for i in range(num_members)
]
```

```python
agent = TinyAgent(
        name="main_agent",
        model=_MAIN_AGENT_MODEL,
        output_root="notebooks/output",
        subagents=web_research_agents,
        **{**_MAIN_AGENT_MODEL_CONFIG, **_PROVIDER_CONFIG},
    )
try:
    result = agent(contents="""
Task to analyze the current situation of xAI and recent events. 
Please analyze from various perspectives, especially the team's circumstances and the product's reputation. 
Analyze the personnel structure and composition of their AI team, including historical and current talent quality assessments.
Provide predictions on what xAI's future development direction will be, particularly regarding the team and the product.

Do the task parallelly with different work sources and you must finally provide a report from all the work sources.
**You don't do** any research yourself; instead, you must merge the results from the different research sources.
Read the result file (if it exists) or memory file (if the result does not exist but memory exists) for each topic after the research.
Record all work sources in your memory to use them later for generating the final report.

Notice:
- Avoid striving for excessive perfection, and focus on being concise. 
- Do not feel obligated to cover all topics; stop when you feel you have enough information.
- If a work source cannot be researched and yields no result or memory, skip that work source.

Language restrictions: 
Provde a final report in Simplified Chinese.

Output:
- You only provide **one file** as the final report, please avoid providing multiple files.
- Save the final report to a file at the path {agent.output_location}/result.md".

Response:
Read out the final report file as the response to the user.
""")
finally:
    AgentManager().unregister(agent.agent_id)
```
